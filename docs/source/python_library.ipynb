# Python library

The python module supports three default modes depending on which sets of molecules are available to check against. The three modes are:
1. **redock**: takes as input a predicted molecule (docked ligand), a true molecule (crystal ligand), and a conditioning molcule (crystal protein) 
2. **dock**: takes as input a predicted molecule (docked ligand) and a conditioning molcule (crystal protein)
3. **mol**: takes as input a predicted molecule (docked ligand)

```python
from molbuster import MolBuster
```

### bust redock

This mode expects as inputs the predicted molecule (docked ligand), true molecule (crystal ligand), and the conditioning molcule (crystal protein). 

```python
from molbuster.datasets import four_redocks
four_redocks = four_redocks()
four_redocks
```

```python
db = MolBuster(config="redock")
db.bust_table(four_redocks)
```

```python
db.results
```

```python
db.results_detail
```
