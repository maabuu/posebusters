# Python library

The Python module exposes more results than the command line tool.

```python
from molbusters import MolBusters
```

The sample dataset contains four redocks for four protein-ligand complexes.

```python
from molbusters.datasets import four_redocks
four_redocks = four_redocks()
four_redocks
```

```python
# three default configurations are available: "redock" for redocking, "dock" for docking and "mol" for molecular generation
db = MolBusters(config="redock")
db.bust_table(four_redocks)
db.results
```
