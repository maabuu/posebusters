```python
from posebusters import PoseBusters
from posebusters.cli import _dataframe_from_output
import pandas as pd
```

```python
sdf_file = "inputs/generated_molecules.sdf"
```

```python
# get the binary test report columns
buster = PoseBusters(config="mol")
df = pd.concat(_dataframe_from_output(result, buster.config) for result in buster.bust(sdf_file, None, None))
df
```

```python
# get the full report 
buster = PoseBusters(config="mol")
df = pd.concat(_dataframe_from_output(result, buster.config, True) for result in buster.bust(sdf_file, None, None))
df
```
